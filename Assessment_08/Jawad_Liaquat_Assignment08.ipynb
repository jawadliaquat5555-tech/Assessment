```python
#Task-1
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "Fail"
        
print(calculate_grade(95))
print(calculate_grade(82))
print(calculate_grade(74))
print(calculate_grade(60))
```

```python
#Task-2
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False
        
print(is_even(4))
print(is_even(7))
```

```python
#Task-3
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result = result * i
    return result

print(factorial(3))
print(factorial(5))
```

```python
#Task-4
def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(find_max(10, 20, 15))
print(find_max(55, 33, 77))
print(find_max(5, 5, 2))
```

```python
#Task-5
def count_words(sentence):
    words = sentence.split()  
    return len(words)         

# Function call
sentence = "I love Python programming"
print(count_words(sentence))
```

```python
#Task-6
def is_prime(n):
    if n <= 1:
        return False 

    for i in range(2, n):
        if n % i == 0:
            return False  
    return True  


print(is_prime(2))   
print(is_prime(9))   
print(is_prime(17))  
```

```python
#Task-7
def list_sum(numbers):
    total = 0              
    for num in numbers:    
        total += num      
    return total          

# Function call
nums = [10, 20, 30]
print(list_sum(nums))
```

```python
#Task-8
def check_password(password):
    if len(password) < 8:
        return "Weak"

    has_digit = False
    for char in password:
        if char.isdigit():  
            has_digit = True
            break

    if not has_digit:
        return "Weak"

    return "Strong"

print(check_password("password123"))  
print(check_password("pass"))          
print(check_password("mypassword"))  
print(check_password("abc12345"))     
```

```python
#Task-9
def celsius_to_fahrenheit(c):
    f = (c * 9/5) + 32
    return f

# Function calls
print(celsius_to_fahrenheit(0))    
print(celsius_to_fahrenheit(100))  
print(celsius_to_fahrenheit(37))   
```
